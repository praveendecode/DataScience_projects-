**Step 3: Data Manipulation:**

````python
# Assigning new Dataframe..

df1=df
````

````python
df1=df[['age', 'status', 'sex', 'orientation', 'body_type', 'diet', 'drinks',
       'drugs', 'education', 'ethnicity', 'height', 'income', 'job',
       'last_online', 'location', 'offspring', 'pets', 'religion', 'sign',
       'smokes', 'speaks']]
````

````python
df1.columns=['Age', 'Status', 'Sex', 'Orientation', 'Body_type', 'Diet', 'Drinks',
       'Drugs', 'Education', 'Ethnicity', 'Height', 'Income', 'Job',
       'Date(Last seen)', 'Location', 'Offspring', 'Pets', 'Religion', 'Zodiac_Sign',
       'Smoke_Habit', 'Language_known']
````

**Datetime format:**

````python
# change date col into datetime format 

df1["Date(Last seen)"] = df1["Date(Last seen)"].str[0:10]     #2012-06-28 /-20-30

df1["Date(Last seen)"] = pd.to_datetime(df1["Date(Last seen)"])

df1["Year"]=df["last_online"].str[0:4]


# Note: 

# if you got any error while running this cell dont worry just run  cells  from (step 3)  again it will work.
````

**Numeric format:**

````python
df1["Age"]       = pd.to_numeric(df1["Age"])
df1["Height"]    = pd.to_numeric(df1["Height"])
df1["Income"]    = pd.to_numeric(df1["Income"])
````

**Step 4: Handling missing values and duplicates:**

````python
# Get count of null vaulues 

print( f"Total Null Values of Each Columns are :\n\n{df1.isnull().sum()}")
````

**Replacing Null values :** 

````python
# Totally 14 columns has Null values 

df1['Body_type'].fillna("Not Mention",inplace=True)

df1['Diet'].fillna("Not Mention",inplace=True)

df1['Drinks'].fillna("Not Mention",inplace=True)

df1['Drugs'].fillna("Not Mention",inplace=True)

df1['Education'].fillna("Not Mention",inplace=True)

df1['Ethnicity'].fillna("Not Mention",inplace=True) 

df1['Height'].fillna(method="pad",inplace=True) 

df1['Job'].fillna("Not Mention",inplace=True) 

df1['Offspring'].fillna("Not Mention",inplace=True) 

df1['Pets'].fillna("Not Mention",inplace=True) 

df1['Religion'].fillna("Not Mention",inplace=True) 

df1['Zodiac_Sign'].fillna("Not Mention",inplace=True) 

df1['Smoke_Habit'].fillna("Not Mention",inplace=True) 

df1['Language_known'].fillna("Not Mention",inplace=True) 

df1=df1.reset_index(drop=True).copy()

````

## **Insights:**


            

**Data_Visualization :**

````python
# Answer 1:

# code:

plt.figure(figsize=(19,5))
df1.query("Status=='single'").groupby("Age").size().plot(kind="bar")
plt.xlabel("Age Groups")
plt.ylabel("Count")
plt.xticks(rotation="horizontal")
plt.grid(color="k",linewidth=0.2,linestyle="--")
plt.title("The Age Group has Maximun singles ")
plt.show()

###################################################### insights ######################################################

      #  most of singles are in age groups Between 24 to 30 .
      #  It gradually goes up from  age(22) then gradually goes downs at age (32).
````

````python
# Anwser 2 :

# code : 

plt.figure(figsize=(19,6))

sns.barplot(x="Status",y="Age",hue="Sex",data=df1,ci=None)

plt.xlabel("Gender{M:F}",labelpad=30)

plt.ylabel("Age Groups",labelpad=20)

plt.grid(color="k",linewidth=0.2)

plt.title("The Status of  Both Gender  In Each Age Group")

plt.show()

###################################################### insights ######################################################
  
    # Compared to male , female singles persons are maximun in the age above 30....

    
````

````python
# Answer 3:

# code :
plt.figure(figsize=(21,5))
df1.query("Status=='single'").groupby("Location").size().sort_values(ascending=False).head(10).plot(kind='bar',color="pink")
plt.title("Top  10  Location  Has  Maximun  Singles ")
plt.xlabel("Location",labelpad=30)
plt.xticks(rotation="horizontal")
plt.grid(linewidth=0.9)
plt.ylabel("Count",labelpad=20)
plt.show()

###################################################### insights ######################################################

      # san francisco, california has maximum singles when compared to other cities .

````

````python
# Answer 4: 

# code : 

plt.figure(figsize=(5,5))
df1.query("Status=='single'").groupby("Year").size().sort_values(ascending=False).plot(kind="bar",color="red")
plt.title(" The Year has Maximun Singles")
plt.xlabel("Year",labelpad=20)
plt.xticks(rotation="horizontal")
plt.ylabel("Count",labelpad=20)
plt.show()

###################################################### insights ######################################################

        # In The year 2012 we had maximum singles 

        
````

````python
# Answer 5: How many persons  avalilable for relationships (Both male & female) ?

# code : 
plt.figure(figsize=(5,5))
df1.query("Status=='available'").groupby("Status").size().plot(kind='bar')
plt.title("Available Persons Count for Relationships")
plt.xlabel("Persons",labelpad=20)
plt.xticks(rotation="horizontal")
plt.ylabel("Count",labelpad=20)
plt.grid(linewidth=0.4,color="black")
plt.show()


###################################################### insights ######################################################

        # The count of persons available for relationships is  1865 
        
````
